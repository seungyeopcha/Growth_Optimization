```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.image as img
import os
import glob
import time
import sklearn
from sklearn.linear_model import LinearRegression
import pickle
```

```python
#이미지->면적검출 함수 정의
def from_image_get_length(path):
    green_Lower=np.array([40,40,40])
    green_Upper=np.array([80,255,255])
    total_weight=0
  
    #path 가 있으면
    img = cv2.imread(path , cv2.IMREAD_COLOR)
    img = cv2.resize(img , None ,  fx = 0.2 , fy = 0.2 ,interpolation = cv2.INTER_AREA)

    #BGR to HSV
    img_hsv= cv2.cvtColor(img , cv2.COLOR_BGR2HSV)
    #maskt
    img_mask = cv2.inRange(img_hsv , green_Lower , green_Upper)
    #contour
    contours_green, _ = cv2.findContours(img_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    #면적검출
    for contour in contours_green:
        total_weight +=cv2.contourArea(contour)
    return total_weight
```
