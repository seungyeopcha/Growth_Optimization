```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
import pickle
```

## 본격적으로 딕셔너리에 내용(면적 , 기존 라벨 등등) 넣고 파일 반복적으로 만드는 과정

```python
#필요한 method 정리
#dict형태로 만드는거(결과를 넣을)
def make_result_dict():
    total_result = dict()
    case_list = sorted(list(os.listdir('C:/data/dacon/open/train')))
    #sorted_list = [x for x in sorted_list if x[0] =="C" ]
    
    for i in case_list:
        total_result[i] = dict()
        total_result[i]["image_area_green"] = list()
        total_result[i]["image_area_yellow"] = list()
        total_result[i]["label"] = list()
     
    return total_result

#green 면적
def from_image_green_area(path):
    green_Lower=np.array([35,15,15])
    green_Upper=np.array([80,255,255])
    
    total_area = 0
  
    #path 가 있으면
    img = cv2.imread(path , cv2.IMREAD_COLOR)
    img = cv2.resize(img , None ,  fx = 0.2 , fy = 0.2 ,interpolation = cv2.INTER_AREA)
 
    #BGR to HSV
    img_hsv= cv2.cvtColor(img , cv2.COLOR_BGR2HSV) #BGR_TOHSV
    img_mask = cv2.inRange(img_hsv , green_Lower , green_Upper)
    countours_green , _= cv2.findContours(img_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    #면적
    for countour in countours_green:
        total_area+= cv2.moments(countour)["m00"]
    return total_area

#yellow면적
def from_image_yellow_area(path):
    yellow_Lower=np.array([10,40,40])
    yellow_Upper=np.array([33,255,255])
    total_area= 0
  
    #path 가 있으면
    img = cv2.imread(path , cv2.IMREAD_COLOR)
    img = cv2.resize(img , None ,  fx = 0.2 , fy = 0.2 ,interpolation = cv2.INTER_AREA)
    #BGR to HSV
    img_hsv= cv2.cvtColor(img , cv2.COLOR_BGR2HSV) #BGR_TOHSV
    img_mask = cv2.inRange(img_hsv , yellow_Lower , yellow_Upper)
    countours_yellow , _= cv2.findContours(img_mask , cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    #면적
    for countour in countours_yellow:
        total_area += cv2.moments(countour)["m00"]
    return total_area
```

```python
total_result = make_result_dict() #틀 만듬
print(total_result)
```

```python
train_path ='C:/data/dacon/open/train/'
index =0
for case in total_result.keys():
    
    #df 값 가져와서 넣기
    case_path = train_path + str(case) + "/label.csv"
    df = pd.read_csv(case_path)
    total_result[case]["label"] = list(df["leaf_weight"])
  
    #이미지 경로로 들어가는 부분
    image_list = sorted(list(os.listdir(train_path+ case +"/image")))
    
    for index , image in enumerate(image_list):
      
      #이미지 경로만 잡아주면 됨
        image_path = str()
        image_path = train_path + str(case) +"/image/" + str(image)
        
        #면적 구하는 메소드 리턴값을 면적
        total_result[case]["image_area_green"].append(from_image_green_area(image_path))
        total_result[case]["image_area_yellow"].append(from_image_yellow_area(image_path))
    print(f"{case} is totally clear")

print("finished")
```

## 병합하기 전에 df에 넣어주고 병합
- case01_... 에다 면적이랑 라벨 그리고 케이스 이름들을 넣어주는 코드 

```python
#train >> case01 >> meta >> CASE0{}_0{}들어가서 merge
#train 인덱스 enumerate하면 0으로 들어감

train_path ='C:/data/dacon/open/train/'
train_list = os.listdir(train_path)
sorted_list = sorted(train_list) #기존 train list의 path에는 sorting 이 안되어 있었는데 sorted 함수를 이용하여 sorting 완료
sorted_list = [x for x in sorted_list if x[0] =="C" ]
for case_index , case_path in enumerate(sorted_list):
  #1번째는 case_index는 0 그리고 case_path는 "CASE01"
  #case path에서 meta까지 들어가야 반복문 가능
    case_path_total = str(train_path) + str(case_path) + "/meta/"
  
  #meta까지 들어갈 수 있는 case_path 가져옴
    meta_list = os.listdir(case_path_total)
  #input output 경로 설정 
    input_file = case_path_total
    output_file = str(case_path_total) + f"merge_case0{case_index+1}.csv" #출력 파일 경로 설정 완료
  
    all_File_list = glob.glob(os.path.join(input_file , "CASE*"))
  
    alldata = [] #읽어 들인 csv 파일 내용 저장할 리스트 만듬
    for index , file in enumerate(all_File_list):
    
        df = pd.read_csv(file)

    #여기에서 outlier 처리 하고 case 끝나면 병합
        df["case"] = str(case_path[-2:])
        df["image_area_green"] = total_result[case_path]["image_area_green"][index]
        df["image_area_yellow"] = total_result[case_path]["image_area_yellow"][index]
        df["무게"] = total_result[case_path]["label"][index]
        alldata.append(df) #빈 리스트에 다 추가
    
  
  
    datacombine = pd.concat(alldata , axis = 0 , ignore_index = True) #concat을 사용하여 
    datacombine.to_csv(output_file , index = False) #csv로 저장
    print(case_index+1 , "is completed")
```
